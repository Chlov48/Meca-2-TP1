```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.stats import linregress
import scienceplots

TITRE = "fig 1 : périodes du pendule en fonction de Φ"

ABSICCES = "cos(Φ)"
uniteA = "radians"
ORDONNEE = "Périodes"
uniteB = "s"

M = np.loadtxt("mesures2.txt",delimiter=",",skiprows=1)
O = np.loadtxt("mesures2.txt",delimiter=",")*np.pi/180
N = np.sum(M,0)/3

plt.plot(np.sqrt(1/np.cos(O[0,:])),N[:],".",color = "black",label="Tₑₓₚ")

mpl.rcParams['figure.figsize'] = [6, 4]

plt.ylabel(ORDONNEE+" (en "+uniteB+")")
plt.xlabel(ABSICCES)
plt.title(TITRE)

plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.stats import linregress
import scienceplots

TITRE = "fig 2 : modèle du pendule simple"

ABSICCES = "cos(Φ)"
uniteA = "radians"
ORDONNEE = "Périodes"
uniteB = "s"

X2 = np.linspace(1,1.45,100)
X = np.linspace(0,1.074,100)

def g2(x):
    return (1.165879888248231 * x + -0.15567354316116866)

def f(x):
    return 2*np.pi*np.sqrt(.2815/(9.81*np.cos(x)))

M = np.loadtxt("mesures2.txt",delimiter=",",skiprows=1)
O = np.loadtxt("mesures2.txt",delimiter=",")*np.pi/180
N = np.sum(M,0)/3

plt.plot(np.sqrt(1/np.cos(O[0,:])),N[:],".",color = "black",label="Tₑₓₚ")
plt.plot(X2,g2(X2),'--',color="black",linewidth=1, label="Tₑₓₚ")
plt.plot(np.sqrt(1/np.cos(X)),f(X),linestyle='dashdot',color="blue",linewidth=1, label="T(Φ)")

mpl.rcParams['figure.figsize'] = [6, 4]

plt.ylabel(ORDONNEE+" (en "+uniteB+")")
plt.xlabel(ABSICCES)
plt.title(TITRE)

plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.stats import linregress
import scienceplots

TITRE = "fig 3 : incertitudes des modèles"

ABSICCES = "cos(Φ)"
uniteA = "radians"
ORDONNEE = "Périodes"
uniteB = "s"

X2 = np.linspace(1,1.45,100)
X = np.linspace(0,1.074,100)

xmed = linregress(np.sqrt(1/np.cos(O[0,:])),N[:])

def g(x):
    return (xmed[0] * x + xmed[1])

def gmin(x):
    return ((xmed[0]+xmed[3]) * x + xmed[1]+xmed[4])

def gmax(x):
    return ((xmed[0]-xmed[3]) * x + xmed[1]-xmed[4])

def f(x):
    return 2*np.pi*np.sqrt(.2815/(9.81*np.cos(x)))

M = np.loadtxt("mesures2.txt",delimiter=",",skiprows=1)
O = np.loadtxt("mesures2.txt",delimiter=",")*np.pi/180
N = np.sum(M,0)/3

plt.plot(np.sqrt(1/np.cos(O[0,:])),N[:],".",color = "black",label="Tₑₓₚ")

plt.plot(X2,g(X2),'--',color="black",linewidth=1, label="Tₑₓₚ")

plt.plot(X2,gmin(X2),'-',color="black",linewidth=0.3)
plt.plot(X2,gmax(X2),'-',color="black",linewidth=0.3)

plt.plot(np.sqrt(1/np.cos(X)),f(X),linestyle='dashdot',color="blue",linewidth=1, label="T(Φ)")

plt.plot(np.sqrt(1/np.cos(X)),f(X)+0.008,color="blue",linewidth=0.3)
plt.plot(np.sqrt(1/np.cos(X)),f(X)-0.008,color="blue",linewidth=0.3)

mpl.rcParams['figure.figsize'] = [6, 4]

plt.ylabel(ORDONNEE+" (en "+uniteB+")")
plt.xlabel(ABSICCES)
plt.title(TITRE)

plt.legend()
```
